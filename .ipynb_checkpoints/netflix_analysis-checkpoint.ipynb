```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")

df.head()
```

```python
df.info()
df.describe()
```

```python
df['country'] = df['country'].fillna("Unknown")
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Unknown")
```

```python
#Movies VS TV shows
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows")
plt.show()
```

```python
#Top Countries
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top Countries")
plt.show()
```

```python
#Content Growth
df['release_year'].value_counts().sort_index().plot()
plt.title("Content Growth Over Years")
plt.show()
```

```python
#Ratings
df['rating'].value_counts().plot(kind='bar')
plt.title("Rating Distribution")
plt.show()
```

```python
#Genre
df['listed_in'].value_counts().head(10).plot(kind='bar')
plt.title("Top Genres")
plt.show()
```

```python
#Movie Duration
movies = df[df['type'] == "Movie"]

movies['duration'].value_counts().head(10).plot(kind='bar')

plt.title("Most Common Movie Durations")

plt.show()
```

```python
#Tv shows Season Analysis
shows = df[df['type'] == "TV Show"]

shows['duration'].value_counts().head(10).plot(kind='bar')

plt.title("Most Common Number of Seasons")

plt.show()
```

```python
#Content Added Per Year
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df['year_added'] = df['date_added'].dt.year

df['year_added'].value_counts().sort_index().plot(figsize=(10,5))

plt.title("Content Added to Netflix Per Year")

plt.show()
```
